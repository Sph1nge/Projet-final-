```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données à partir des bases de données externes
data_2018 = pd.read_csv('MissingMigrants-Global-2018.csv')
data_2019 = pd.read_csv('MissingMigrants-Global-2019.csv')
data_2020 = pd.read_csv('MissingMigrants-Global-2020.csv')
data_2021 = pd.read_csv('MissingMigrants-Global-2021.csv')
data_2022 = pd.read_csv('MissingMigrants-Global-2022.csv')

# Exemple de graphique circulaire pour chaque année
fig, axs = plt.subplots(1, 5, figsize=(50, 100))

sizes_2018 = data_2018.groupby('Region')['Total Number of Dead and Missing'].sum()
axs[0].pie(sizes_2018, labels=sizes_2018.index, autopct='%1.1f%%')
axs[0].set_title('Répartition des migrations en 2018')

sizes_2019 = data_2019.groupby('Region')['Total Number of Dead and Missing'].sum()
axs[1].pie(sizes_2019, labels=sizes_2019.index, autopct='%1.1f%%')
axs[1].set_title('Répartition des migrations en 2019')

sizes_2020 = data_2020.groupby('Region')['Total Number of Dead and Missing'].sum()
axs[2].pie(sizes_2020, labels=sizes_2020.index, autopct='%1.1f%%')
axs[2].set_title('Répartition des migrations en 2020')

sizes_2021 = data_2021.groupby('Region')['Total Number of Dead and Missing'].sum()
axs[3].pie(sizes_2021, labels=sizes_2021.index, autopct='%1.1f%%')
axs[3].set_title('Répartition des migrations en 2021')

sizes_2022 = data_2022.groupby('Region')['Total Number of Dead and Missing'].sum()
axs[4].pie(sizes_2022, labels=sizes_2022.index, autopct='%1.1f%%')
axs[4].set_title('Répartition des migrations en 2022')

plt.tight_layout()
plt.show()


# Exemple d'histogramme
plt.figure(figsize=(8, 6))
plt.bar(data_2018['Region'], data_2018['Total Number of Dead and Missing'], label='2018')
plt.bar(data_2019['Region'], data_2019['Total Number of Dead and Missing'], label='2019')
plt.bar(data_2020['Region'], data_2020['Total Number of Dead and Missing'], label='2020')
plt.bar(data_2021['Region'], data_2021['Total Number of Dead and Missing'], label='2021')
plt.bar(data_2022['Region'], data_2022['Total Number of Dead and Missing'], label='2022')
plt.xlabel('Region')
plt.ylabel('Total Number of Dead and Missing')
plt.title('Migrations annuelles')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Exemple de diagramme
plt.figure(figsize=(12, 6))
plt.plot(grouped_2018.index, grouped_2018.values, marker='o', label='2018')
plt.plot(grouped_2019.index, grouped_2019.values, marker='o', label='2019')
plt.plot(grouped_2020.index, grouped_2020.values, marker='o', label='2020')
plt.plot(grouped_2021.index, grouped_2021.values, marker='o', label='2021')
plt.plot(grouped_2022.index, grouped_2022.values, marker='o', label='2022')
plt.xlabel('Region')
plt.ylabel('Total Number of Dead and Missing')
plt.title('Tendances annuelles des migrations')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Graphique baton des mortalitées et disparition sur les flux migratoires
plt.figure(figsize=(10, 6))
plt.bar(data_2018['Region'], data_2018['Total Number of Dead and Missing'], label='Total Number of Dead and Missing')
plt.xlabel('Region')
plt.ylabel('Total Number of Dead and Missing')
plt.title('Étude des mortalitées et disparition sur les flux migratoires')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Exemple de graphique en bâtons groupés
plt.figure(figsize=(10, 6))
bar_width = 0.15
index = range(len(data_2018['Region']))
plt.bar(index, data_2018['Total Number of Dead and Missing'], width=bar_width, label='2018')
plt.bar([i + bar_width for i in index], data_2019['Total Number of Dead and Missing'], width=bar_width, label='2019')
plt.bar([i + 2 * bar_width for i in index], data_2020['Total Number of Dead and Missing'], width=bar_width, label='2020')
plt.bar([i + 3 * bar_width for i in index], data_2021['Total Number of Dead and Missing'], width=bar_width, label='2021')
plt.bar([i + 4 * bar_width for i in index], data_2022['Total Number of Dead and Missing'], width=bar_width, label='2022')
plt.xlabel('Region')
plt.ylabel('Total Number of Dead and Missing')
plt.title('Migrations annuelles (Bâtons groupés)')
plt.xticks([i + 2 * bar_width for i in index], data_2018['Region'], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Exemple de graphique en aires empilées
plt.figure(figsize=(10, 6))
plt.stackplot(data_2018['Region'], data_2018['Total Number of Dead and Missing'],
              data_2019['Total Number of Dead and Missing'],
              data_2020['Total Number of Dead and Missing'],
              data_2021['Total Number of Dead and Missing'],
              data_2022['Total Number of Dead and Missing'],
              labels=['2018', '2019', '2020', '2021', '2022'])
plt.xlabel('Region')
plt.ylabel('Total Number of Dead and Missing')
plt.title('Migrations annuelles (Aires empilées)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Exemple de radar chart
categories = data_2018['Region']
values_2018 = data_2018['Total Number of Dead and Missing']
values_2019 = data_2019['Total Number of Dead and Missing']
values_2020 = data_2020['Total Number of Dead and Missing']
values_2021 = data_2021['Total Number of Dead and Missing']
values_2022 = data_2022['Total Number of Dead and Missing']

plt.figure(figsize=(8, 8))
plt.plot(categories, values_2018, marker='o', label='2018')
plt.plot(categories, values_2019, marker='o', label='2019')
plt.plot(categories, values_2020, marker='o', label='2020')
plt.plot(categories, values_2021, marker='o', label='2021')
plt.plot(categories, values_2022, marker='o', label='2022')
plt.fill(categories, values_2018, alpha=0.25)
plt.fill(categories, values_2019, alpha=0.25)
plt.fill(categories, values_2020, alpha=0.25)
plt.fill(categories, values_2021, alpha=0.25)
plt.fill(categories, values_2022, alpha=0.25)
plt.xlabel('Region')
plt.ylabel('Total Number of Dead and Missing')
plt.title('Tendances annuelles des migrations (Radar Chart)')
plt.legend()
plt.tight_layout()
plt.show()
```
